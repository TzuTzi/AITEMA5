```python
import pandas as pd
import numpy as np
import torch
from sklearn import linear_model

df = pd.read_csv("data/v3_world-happiness-report-2017.csv", low_memory=False)
df = df.dropna(subset=["Family", "Happiness.Score", "Economy..GDP.per.Capita.", "Freedom"])

fam = df["Family"].tolist()
hapiness = df["Happiness.Score"].tolist()
gdp = df["Economy..GDP.per.Capita."].tolist()
freedom = df["Freedom"].tolist()

np.random.seed(47)
indexes = [i for i in range(len(hapiness))]
trainSample = np.random.choice(indexes, int(0.8 * len(hapiness)), replace=False)
validationSample = [i for i in indexes if not i in trainSample]

trainInputsFamily = [fam[i] for i in trainSample]
trainOutputsHapiness = [hapiness[i] for i in trainSample]

trainInputsGdp = [gdp[i] for i in trainSample]
trainInputsFreedom = [freedom[i] for i in trainSample]
trainOutputs = [hapiness[i] for i in trainSample]

validationInputsFamily = [fam[i] for i in validationSample]
validationInputsGdp = [gdp[i] for i in validationSample]
validationInputsFreedom = [freedom[i] for i in validationSample]
validationOutputs = [hapiness[i] for i in validationSample]

# Model 1: Happiness ~ Family
xx = [[el] for el in trainInputsFamily]
regressor = linear_model.LinearRegression()
regressor.fit(xx, trainOutputsHapiness)
w0, w1 = regressor.intercept_, regressor.coef_
print("weights: ")
print(w0)
print(w1)

# Model 2: Happiness ~ GDP + Freedom
xx = [[el1, el2] for el1, el2 in zip(trainInputsGdp, trainInputsFreedom)]
regressor = linear_model.LinearRegression()
regressor.fit(xx, trainOutputs)
w0, w1 = regressor.intercept_, regressor.coef_
print("weights: ")
print(w0)
print(w1)
```
