```python
from operator import countOf
from random import random

import isOdd
from scipy.linalg import inv


class SuperDuperWowLinearInstant:
    def __init__(self):
        self.w1=random()
        self.w2=random()
    def update(self,x,y):
        self.w1= (len(x) * self.suma_produs(x, y) - self.suma(x) * self.suma(y)) / (len(x) * self.suma_produs(x, x) - sum(x) ** 2)
        self.w2=(self.suma(y)-self.w1*self.suma(x))/len(x)

    def suma_produs(self,x,y):
        suma=0
        for x,y in zip(x,y):
            suma+=x*y
        return suma
    def suma(self,x):
        suma=0
        for x in x:
            suma+=x
        return suma
```

```python
import numpy as np
from scipy import linalg
df = pd.read_csv("data/v1_world-happiness-report-2017.csv",low_memory=False);
gdp=df["Economy..GDP.per.Capita."]
freedom=df["Freedom"]
hapiness=df["Happiness.Score"]
class LinearWith2:
    def __init__(self):
        self.weights = None

    def matrix(self, x1, x2):
        n = len(x1)
        ones = np.ones((n, 1))
        self.X = np.hstack([ones, np.array(x1).reshape(-1, 1), np.array(x2).reshape(-1, 1)])

    def weights_matrix(self, y):
        y = np.array(y).reshape(-1, 1)
        self.weights = linalg.inv(self.X.T @ self.X) @ self.X.T @ y

    def compute(self, x):
        rez=float((self.weights[0] + self.weights[1] * x[0] + self.weights[2] * x[1]).item())
        return rez
l1=LinearWith2()
l1.matrix(gdp,freedom)
l1.weights_matrix(hapiness)
print(l1.compute([gdp[1],freedom[1]]))
```
